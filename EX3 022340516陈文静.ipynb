```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams["font.family"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False  # 

# 1. 
dates = [3, 8, 13, 17, 25, 31]  # 7
rate_2017 = [6.8007, 6.8060, 6.7835, 6.7700, 6.7511, 6.7265]  # 2017
rate_2019 = [6.8640, 6.8881, 6.8662, 6.8827, 6.8737, 6.8841]  # 2019

# 2. 
plt.figure(figsize=(10, 6))  # 

# 20171
plt.plot(dates, rate_2017, 
         linestyle='-.',  # 
         color='g',       # 
         linewidth=1,     # 1
         label='20177')

# 2019RGB3
plt.plot(dates, rate_2019, 
         linestyle='--',  # 
         color=(0.8, 0.2, 0.4),  # RGB
         linewidth=3,     # 3
         label='20197')

# 3. 
plt.title('2017720197/', fontsize=12)
plt.xlabel('7', fontsize=10)
plt.ylabel('', fontsize=10)
plt.legend(fontsize=10)  # 
plt.grid(linestyle=':', alpha=0.5)  # 

# 4. 
plt.tight_layout()  # 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams["font.family"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False

# 1. A/C
quarters = ['1', '2', '3', '4']
sales_A = [2144, 4617, 7674, 4409]  # A
sales_C = [153, 1214, 2414, 680]    # C

# 2. 
plt.figure(figsize=(10, 6))

# A
plt.plot(quarters, sales_A, 
         marker='o',  # 
         color='teal', 
         linewidth=2,
         label='A')

# C
plt.plot(quarters, sales_C, 
         marker='o',  # 
         color='purple', 
         linewidth=2,
         label='C')

# ACHTML#ADD8E60.3
plt.fill_between(quarters, sales_A, sales_C,
                 facecolor='#ADD8E6',  # HTML
                 alpha=0.3,            # 0.3
                 label='A/C')

# 3. 
plt.title('AC', fontsize=12)
plt.ylabel('', fontsize=10)
plt.legend(fontsize=10)
plt.grid(linestyle=':', alpha=0.5)

# 4. 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams["font.family"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False

# 1. 
np.random.seed(42)  # 
x = np.random.rand(50) * 10  # 50x0-10
y = np.random.rand(50) * 10  # 50y0-10
color_values = np.random.rand(50)  # 500-1

# 2. 
plt.figure(figsize=(10, 6))

# Blues100
scatter = plt.scatter(x, y,
                     c=color_values,    # 
                     cmap='Blues',      # SequentialBlues
                     s=100,             # 100
                     alpha=0.7)         # 

# 3. 
plt.colorbar(scatter, label='')  # 
plt.title('Blues', fontsize=12)
plt.xlabel('X', fontsize=10)
plt.ylabel('Y', fontsize=10)
plt.grid(linestyle=':', alpha=0.5)

# 4. 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False  # 

# 1. 
dates = np.arange(4, 19)  # 418
max_temp = [28, 30, 32, 34, 31, 29, 27, 28, 30, 31, 29, 28, 27, 26, 25]  # 
min_temp = [18, 20, 22, 23, 21, 19, 17, 18, 20, 21, 19, 18, 17, 16, 15]  # 

# 2. 
plt.figure(figsize=(12, 6))

# 
plt.plot(dates, max_temp, 
         marker='^', mec='red', mfc='white', markersize=8, 
         color='red', linewidth=2, label='')

# 
plt.plot(dates, min_temp, 
         marker='v', color='blue', linewidth=2, label='')

# 3. 347
max_idx = max_temp.index(34)
max_date = dates[max_idx]
plt.annotate(
    '', xy=(max_date, 34), 
    xytext=(max_date+1, 35), 
    arrowprops=dict(arrowstyle='->', color='red'), 
    fontsize=10
)

# 4. 
plt.title('', fontsize=12)
plt.xlabel('', fontsize=10)
plt.ylabel('', fontsize=10)
plt.legend(fontsize=10)
plt.grid(linestyle=':', alpha=0.5)
plt.xticks(dates)  # 

# 5. 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["mathtext.fontset"] = "cm"

plt.style.use('ggplot')
x = np.linspace(0, 5, 100)
y = x ** 2

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='darkblue', linewidth=2, label=r'$y = x^2$')
plt.title('ggploty=x', fontsize=12, color='black')
plt.xlabel('X', fontsize=10, color='black')
plt.ylabel('Y', fontsize=10, color='black')
plt.legend(fontsize=10, prop={'weight': 'normal', 'size': 10})
plt.tight_layout()
plt.show()
```
